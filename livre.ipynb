   ## Rappel a la demande:
   
   L'etude consiste a trouver une reponse a l'appel de l'ADEME (Agence de L'Environnement et de la Maîtrise de l'Énergie)  qui consiste à promouvoir la réalisation de démonstrateurs et d’expérimentations de nouvelles solutions de mobilité pour les personnes et les marchandises adaptées à différents types de territoires.Les nouvelles technologies de transport sont certes plus économiques et moins polluantes mais il reste ces problèmes de logistique du transport qui présentent un enjeu majeur pour l’avenir et leur impact sur l’environnement peut être véritablement significatif et cette problématique se déploit de plusieurs manières : distribution du courrier, livraison de produits, traitement du réseau routier, ramassage des ordures, etc.
   
   ## Objectif:
   
   Le but de cette étude est de générer une tournée de livraison (problème du VRP). Le problème algorithmique consiste donc à calculer sur un réseau routier une tournée permettant de relier entre elles un sous-ensemble de villes, puis de revenir à son point de départ, de manière à minimiser la durée totale de la tournée. Cette optimisation devra tenir compte du trafic prévu sur chaque axe pour les différentes tranches horaires.
   
   ## Nos Contraintes:
   
   Parmi les contraintes proposées dans le sujet du projet, nous avons choisi :
   
   • la contrainte du traffic généré selon une plage horaire.
